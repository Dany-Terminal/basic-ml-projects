```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('dna_sample/synthetic_dna_dataset.csv')
df.head()
```

```python
df.drop(labels='Sample_ID',axis=1,inplace=True)
df.head()
```

```python
df['Class_Label'].unique()
```

```python
df['Class_Label']=df['Class_Label'].map({'Bacteria':1, 'Virus':0, 'Human':3,'Plant':2})
df.head()
```

```python
#Check if stratified is required
df['Disease_Risk'].value_counts(normalize=True)*100
#evenly distributed thus stratified not required. Still use it to prevent unluck random split
```

### Predicting Varible is 'Disease Risk'.

- svm is great for high dimension and textual data however most of the data is numeric 10 dim, SVM have potential but not top priority
- Random Forest Classification is good for complex data and is a combination of multiple decision tree, so it looks good
- knn good for low dimension and low complex data, not good in this case
- decision tree may also perform good , but might not as good as random forest tree, so its on low priority
- Naive bayes works on very high dim where dimension>no. of rows, In this case the DNA might be co-related and naive bayes treats each attribute indivisually for prediction, thus might not be the best, so low priority

### Cross validation will be done on models
1) Random Forest
2) Decision Tree
3) XGBoost
4) ADABoost
5) Gradient Boost

Splitting of dataset needs to be considered.

    Since we are doing classification we will use Stratified split, and also need to use cross-validation
    Thus we will use  'Stratified K-Fold' Splitting

*Note:* We are using all trees since this dataset might be best with it , also the encoding is done using map, which is good for tree based model

```python
from sklearn.model_selection import StratifiedKFold,cross_val_score,RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
```

```python
x=df.drop(columns=['Disease_Risk'])
y=df['Disease_Risk']
```

```python
#Define Models
models={
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random forset": RandomForestClassifier(random_state=42),
    "XGBoost" : xgb.XGBClassifier(eval_metric='mlogloss', random_state=42),
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "Gradient Boost":GradientBoostingClassifier(random_state=42)
}

#For now train all model in default hyper paraneter to see how it looks
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name,model in models.items():
    score=cross_val_score(model,x,y,cv=skf,scoring='accuracy') #Other scoring attributes can also be used
    print(f'{name} :Mean accuracy={score.mean():4f},std={score.std():.4f}')
```

```python
for name,model in models.items():
    score=cross_val_score(model,x,y,cv=skf,scoring='f1_macro') #Other scoring attributes can also be used
    print(f'{name} :Mean f1_macro={score.mean():4f},std={score.std():.4f}')
```

# Selecting Model
- AdaBoost had good accuracy but failed in f1_macro(clearly worse)
- Decision Tree decent but not good
- Gradient Boost, very less std->stable, with good f1macro and accuracy
- XGBoost better model than Gradient in default 
- Random Forest clearly the best model over here in terms of accuracy and f1_score


# We will be choosing 1 model for simplification, practically 2-3 models will be taken and hyper-tunned.

*Note: Having a  accuracy(34%) shows that there might not be a strong co-relation between independent and dependent variables. Thus hyperparameter tunning wont drastically increase the accuracy.*

```python
#Use "GridSearch cv" or "RandomizeSearch cv" for hyper_parameter tunning. Practically Randomize is better since it wont brute force all combination but might miss the best combo
rf=RandomForestClassifier(random_state=42)

parameter_grid={
    'n_estimators': [200,400,600],
    'max_depth':[5,10,15,None],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4],
    'max_features':['sqrt','log2']
}

search= RandomizedSearchCV(rf,parameter_grid,n_iter=30,cv=5,scoring='f1_macro',n_jobs=-1)
```

| Argument              | Meaning                                                                           |
| --------------------- | --------------------------------------------------------------------------------- |
| `rf`                  | The **estimator** (RandomForestClassifier) to tune                                |
| `param_distributions` | The **hyperparameter dictionary** you defined â€” tells it what values to pick from |
| `n_iter=30`           | RandomizedSearchCV will try **30 random combinations** from the dictionary        |
| `cv=5`                | 5-fold cross-validation for each combination                                      |
| `scoring="f1_macro"`  | Use F1-macro to evaluate performance                                              |
| `n_jobs=-1`           | Use all CPU cores for faster computation                                          |
| `random_state=42`     | Makes the random sampling reproducible                                            |


```python
search.fit(x,y)
print(search.best_score_)
print(search.best_params_)
```
